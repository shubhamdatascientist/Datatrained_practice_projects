# Removing the Outliers

```python
#Removing the extreme outliers in XLarge Bags

lower_bridge=df1["XLarge Bags"].quantile(0.25)
upper_bridge=df1["XLarge Bags"].quantile(0.75)
IQR=upper_bridge-lower_bridge
```

```python
upper_boundary=upper_bridge+3*IQR
lower_boundary=lower_bridge-3*IQR
df1.drop(df1[df1["XLarge Bags"]>upper_boundary].index,axis=0,inplace=True)
df1.drop(df1[df1["XLarge Bags"]<lower_boundary].index,axis=0,inplace=True)
```

```python
#Removing the extreme outliers in 4770

lower_bridge=df1["4770"].quantile(0.25)
upper_bridge=df1["4770"].quantile(0.75)
IQR=upper_bridge-lower_bridge
```

```python
upper_boundary=upper_bridge+3*IQR
lower_boundary=lower_bridge-3*IQR
df1.drop(df1[df1["4770"]>upper_boundary].index,axis=0,inplace=True)
df1.drop(df1[df1["4770"]<lower_boundary].index,axis=0,inplace=True)
```

```python
#Removing the extreme outliers in Large Bags

lower_bridge=df1["Large Bags"].quantile(0.25)
upper_bridge=df1["Large Bags"].quantile(0.75)
IQR=upper_bridge-lower_bridge
```

```python
upper_boundary=upper_bridge+3*IQR
lower_boundary=lower_bridge-3*IQR
df1.drop(df1[df1["Large Bags"]>upper_boundary].index,axis=0,inplace=True)
df1.drop(df1[df1["Large Bags"]<lower_boundary].index,axis=0,inplace=True)
```

## Scaling the Dataset

## Splitting the dataset into X and Y

```python
x=df1.drop("AveragePrice",axis=1)
y=df1["AveragePrice"]
```

# Machine Learning Model Training

```python
#Choosing the best random state for our ML Model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    predrf=lr.predict(x_test)
    acc=r2_score(y_test,predrf)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best Accuracy is ",maxAccu," on Random State ",maxRS)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor()
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
acc=r2_score(y_test,predrf)
print("Accuracy Score of Random Forest Regressor: ",acc)
```

```python
#Verifying the performance of the model by graph
import matplotlib.pyplot as plt
plt.figure(figsize=(9,6))
plt.scatter(x=y_test,y=predrf,color='b')
plt.plot(y_test,y_test,color='r')
plt.xlabel("Acutal")
plt.ylabel("Predicted")
plt.title("Random Forest Regression")
plt.show()
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='rbf')
svr.fit(x_train,y_train)
predrf=svr.predict(x_test)
acc=r2_score(y_test,predrf)
print("Accuracy Score of Support Vector Regressor: ",acc)
```

```python
plt.figure(figsize=(9,6))
plt.scatter(x=y_test,y=predrf,color='b')
plt.plot(y_test,y_test,color='r')
plt.xlabel("Acutal")
plt.ylabel("Predicted")
plt.title("Support vector Regression")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
predrf=dtr.predict(x_test)
acc=r2_score(y_test,predrf)
print("Accuracy Score of Decision Tree Regressor: ",acc)
```

```python
plt.figure(figsize=(9,6))
plt.scatter(x=y_test,y=predrf,color='b')
plt.plot(y_test,y_test,color='r')
plt.xlabel("Acutal")
plt.ylabel("Predicted")
plt.title("Support vector Regression")
plt.show()
```
